
###1.
Уравнение имеет вид $y' - y = 2x - 3 $, где $ y' $ обозначает производную функции \( y \) по переменной \( x \).

Это линейное неоднородное дифференциальное уравнение. Для его решения можно использовать метод вариации постоянных или метод интегрирующего множителя.

Шаг 1: Найдем общее решение однородного уравнения

Сначала решим однородное уравнение:

$ y' - y = 0 $

Это уравнение имеет решение:

$y_h = Ce^x $

где $C $ — произвольная константа.

Шаг 2: Найдем частное решение неоднородного уравнения

Теперь найдём частное решение $ y_p $ для неоднородного уравнения:

$ y' - y = 2x - 3 $

Попробуем взять $ y_p = Ax + B $, где $ A $ и $ B $ — константы, которые мы определим.

Теперь найдём производную:

$ y_p' = A $

Подставим $y_p $ и$ y_p' $ в уравнение:

$ A - (Ax + B) = 2x - 3 $

Упрощаем это уравнение:

$ A - Ax - B = 2x - 3 $

Сравним коэффициенты при равных степенях $ x $:

1. При $ x $:
   $ -A = 2 \quad \Rightarrow \quad A = -2 $

2. Свободные члены:
   $ A - B = -3 \quad \Rightarrow \quad -2 - B = -3 \quad \Rightarrow \quad B = -1 $

Таким образом, частное решение:

$ y_p = -2x - 1 $

Шаг 3: Общее решение

Теперь общее решение уравнения будет равно сумме общего решения однородного уравнения и частного решения:

$ y = y_h + y_p = Ce^x - 2x - 1 $

где $ C $ — произвольная константа, определяемая начальными условиями (если они заданы).

Таким образом, общее решение данного уравнения:

$ y = Ce^x - 2x - 1 $

где $ y' $ — это производная функции $ y $ по переменной $ x $, мы можем переписать его в более удобной форме:

$
y' = -\frac{xy + 1}{x^2}.

Это уравнение является линейным обыкновенным дифференциальным уравнением первого порядка.

### Шаг 1: Приведем уравнение к стандартному виду

Запишем уравнение в форме

$
y' + \frac{1}{x}y = -\frac{1}{x^2}.
$

### Шаг 2: Найдем интегрирующий множитель

Интегрирующий множитель $ \mu(x) $ находится по формуле:

$
\mu(x) = e^{\int P(x) \, dx} = e^{\int \frac{1}{x} \, dx} = e^{\ln |x|} = |x|.
$

Для упрощения мы можем просто использовать $\mu(x) = x $.

### Шаг 3: Умножим уравнение на интегрирующий множитель

Теперь умножим всё уравнение на $ x $:

$
x y' + y = -1.
$

### Шаг 4: Преобразуем левую часть уравнения

Левая часть уравнения является производной произведения:

$
\frac{d}{dx}(xy) = -1.
$

### Шаг 5: Интегрируем обе стороны

Теперь интегрируем обе стороны:

$
\int \frac{d}{dx}(xy) \, dx = \int -1 \, dx,
$

что дает:

$
xy = -x + C,
$

где $ C $ — произвольная постоянная.

### Шаг 6: Выразим $ y $

Теперь выразим $ y $:

$
y = -1 + \frac{C}{x}.
$

Таким образом, общее решение данного уравнения:

$
y = -1 + \frac{C}{x},
$

где $ C $ — произвольная константа, которая может быть определена при наличии начальных условий.



### Ряд 1: $ \sum_{n=1}^{\infty} \frac{n+2}{n^2+n+1} $

Чтобы исследовать сходимость ряда, можно использовать признак сравнения или признак Даламбера.

1. **Анализ поведения члена ряда**:

   $
   \frac{n+2}{n^2+n+1} \approx \frac{n}{n^2} = \frac{1}{n} \quad \text{при больших } n
  $

   Сравним с рядом $\sum \frac{1}{n}$, который является гармоническим и расходится.

2. **Признак сравнения**:

   $
   \frac{n+2}{n^2+n+1} > \frac{n}{n^2+n+1} > \frac{n}{n^2+n^2} = \frac{1}{2n}
   $

   Поскольку гармонический ряд $\sum \frac{1}{n}$ расходится, и $\frac{1}{2n}$ расходится, исходный ряд также расходится по признаку сравнения.

### Ряд 2: $ \sum_{n=1}^{\infty} \frac{n^n}{n!} $

Для этого ряда применим признак Даламбера:

1. **Признак Даламбера**:

   $
   \lim_{n \to \infty} \frac{a_{n+1}}{a_n} = \lim_{n \to \infty} \frac{\frac{(n+1)^{n+1}}{(n+1)!}}{\frac{n^n}{n!}}
   $

   Упростим выражение:

   $
   = \lim_{n \to \infty} \frac{(n+1)^{n+1}}{n^n} \cdot \frac{n!}{(n+1)!}
   $

   $
   = \lim_{n \to \infty} \left(\frac{n+1}{n}\right)^n \cdot \frac{n+1}{n+1}
   $

   $
   = \lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n
   $

   $
   = e
   $

   Поскольку $e > 1$, ряд расходится по признаку Даламбера.

### Ряд 3: $ \sum_{n=1}^{\infty} n \left(1 - \frac{1}{n}\right)^{n^2} $

Рассмотрим поведение члена ряда при больших $ n $:

1. **Анализ внутреннего выражения**:

   $
   \left(1 - \frac{1}{n}\right)^{n^2} = \left[\left(1 - \frac{1}{n}\right)^n\right]^n
   $

   При $ n \to \infty $ выражение $\left(1 - \frac{1}{n}\right)^n \to \frac{1}{e}$, тогда:

   $
   = \left(\frac{1}{e}\right)^n
   $

   Член ряда асимптотически ведёт себя как $\frac{n}{e^n}$, который стремится к нулю.

2. **Признак сравнения**:

   Ряд $\sum \frac{n}{e^n}$ сходится по признаку сравнения с геометрическим рядом с $\frac{1}{e} < 1$, следовательно, исходный ряд также сходится.

### Ряд 4: $ \sum_{n=1}^{\infty} \frac{(-1)^{n+1}}{2^n - \ln n} $

Проверим сходимость по признаку Лейбница для знакопеременных рядов:

1. **Члены ряда уменьшаются по модулю**:

   $
   \left|\frac{(-1)^{n+1}}{2^n - \ln n}\right| = \frac{1}{2^n - \ln n}
   $

   При больших $ n $, $ 2^n $ доминирует над $\ln n$, и ряд убывает.

2. **Предел члена ряда при $ n \to \infty $**:

   $
   \lim_{n \to \infty} \frac{1}{2^n - \ln n} = 0
   $

   Таким образом, ряд сходится по признаку Лейбница.

### Ряд 5: $ \sum_{n=1}^{\infty} \frac{(x-2)^{n+1}}{3^n(n+2)} $

Рассмотрим сходимость ряда в зависимости от $ x $:

1. **Признак Даламбера**:

   $
   \lim_{n \to \infty} \left|\frac{a_{n+1}}{a_n}\right| = \lim_{n \to \infty} \left|\frac{(x-2)^{n+2}}{3^{n+1}(n+3)} \cdot \frac{3^n(n+2)}{(x-2)^{n+1}}\right|
   $

   $
   = \lim_{n \to \infty} \left|\frac{(x-2)}{3} \cdot \frac{n+2}{n+3}\right|
   $

   $
   = \left|\frac{x-2}{3}\right| \quad \text{(при \( n \to \infty \), \(\frac{n+2}{n+3} \to 1\))}
   $

   Ряд сходится, если $\left|\frac{x-2}{3}\right| < 1$, что эквивалентно $|x - 2| < 3$. То есть, ряд сходится при $ -1 < x < 5 $.
